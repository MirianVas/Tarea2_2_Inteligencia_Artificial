```python
datos['room_ratio'] = datos['total_rooms'] / datos['total_bedrooms']
```

```python
datos
```

```python
datos['casa_mas_50_aos'] = datos['housing_median_age'].apply(lambda x: 1 if x >= 50 else 0)
```

```python
datos['ingreso_maximo'] = datos['median_income'].apply(lambda x: 1 if x >= 15 else 0)
```

```python
datos['value_por_income'] = datos['median_house_value'] / datos['median_income']
```

```python
datos['rooms_por_households'] = datos['total_rooms'] / datos['households']
```

```python
datos['bedrooms_por_households'] = datos['total_bedrooms'] / datos['households']
```

```python
datos['population_por_households'] = datos['population'] / datos['households']
```

```python
datos.drop(['population'], axis=1, inplace=True)
```

Para aumentar el porcentaje, primero cre nuevas columnas, entre ellas estan:

La columna llamada "casa_mas_50_aos", donde si la casa tiene mas o igual de 50 aos tendr el valor de 1, y si tienes menos tendr valor de 0.

Apliqu el mismo caso para la columna "median_income", cre una nueva columna llamada "ingreso_maximo" donde si es mayor o igual 15 tendr valor de 1, de lo contrario tendr valor de 0.

La relaciones entre las columnas realizadas fueron:

La columna "value_por_income" da una idea de cunto vale una casa en relacin al ingreso medio de las personas en esa rea. Si una casa es mucho ms costosa que el ingreso promedio, podra estar en una zona ms codiciada o con condiciones especficas que elevan el valor.

La columna "rooms_por_households" el promedio de habitaciones por hogar en cada bloque. Este valor ayuda a representar el tamao promedio de las casas en trminos de nmero de habitaciones.Las zonas con ms habitaciones por hogar suelen tener casas ms grandes o ms espaciosas, lo que generalmente incrementa el valor de la casa.

Las columnas "rooms_por_households" y "bedrooms_por_households" calculan el promedio de habitaciones por hogar y dormitorios por hogar respectivamente. El promedio de habitaciones por hogar da una idea del tamao general de las casas y el promedio de dormitorios por hogar es relevante porque muestra la cantidad de dormitorios con los que cuenta una casa.

La columna "population_por_households" calcula el promedio de personas por hogar.

```python
X = datos.drop('median_house_value', axis=1)
y = datos['median_house_value']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

#entrenamiento
modelo.fit(X_train, y_train)
```

```python
predicciones = modelo.predict(X_test)
```

```python
modelo.score(X_test, y_test)
```

```python
modelo.score(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np

mse = mean_squared_error(y_test, predicciones)

mse = np.sqrt(mse)

mse
```

```python
r2_score(y_test, predicciones)
```
